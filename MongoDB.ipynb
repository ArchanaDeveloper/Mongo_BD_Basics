```python
import collections
import pymongo
client = pymongo.MongoClient("mongodb+srv://archanadeveloper16:1234@cluster0.kcrtgwx.mongodb.net/?retryWrites=true&w=majority")
db = client.d6667
collections = db.peopleDetails
d1={"age":38,"name":"Prabu"}
collections.insert_one(d1)
```

```python
a=collections.find( { "age": { "$lte": 40 } } )
for x in a:
 print(x)
```

```python
b=collections.find({"name":{"$regex":"u$"}})
for i in b:
 print(i)
```

```python
a=collections.find( { "age": { "$ne": 40 } } )
for x in a:
 print(x)
```

```python
a=collections.find( { "age": { "$ne": 40 } } ).limit(2)
for x in a:
 print(x)
```

```python
a=collections.find( { "age": { "$gte": 9 },"age":  { "$lt": 100 } } )
for x in a:
 print(x)
```

```python
#update Timestamp
myquery = { "name":"nirmal" }


up= {"$currentDate": { "lastModified": True}}
collections.update_one(myquery,up)
for x in collections.find():
 print(x)
```

```python
#sample data for $inc
data1={
 "_id": 10,
 "name": "Thendral",
 "quantity": 10,
 "metrics": [2, 3.5,"hello"]
}
collections.insert_one(data1)
for x in collections.find():
 print(x)
```

```python
#$increment


myquery={ "_id": 10 }  
update= { "$inc": { "quantity": -2, "metrics.0": 10 } }
collections.update_one(myquery,update)
print(collections.find_one({"_id":10}))
```

```python
#sample data for $multiplication
data2={ "_id" : 101, "item" : "ABC", "price" : 10.99, "qty" : 25 }
#collections.insert_one(data2)
#$mul
myquery1= { "_id": 101 }
update = { "$mul": { "price": 1.25, "qty": 2 } }
collections.update_one(myquery1,update)
for x in collections.find():
 print(x)
```

```python
# rename
collections.update_one({"_id":10},{"$rename":{"quantity":"volume"}})
for x in collections.find():
 print(x)
```

```python
result = collections.insert_many([{"x": 11, "tags1": ["dog", "cat"]},
                               {"x": 22, "tags1": ["mouse", "cat", "dog"]}]
                               )


print(result.inserted_ids)
for x in collections.find():
 print(x)
```

```python
pipeline = [{"$unwind": "$tags1"},{"$sort":{"x":-1}}]
for x in collections.find():
 print(x)
```
